# Experiment 1: Dataset Analysis

## Aim
To analyze the PlantDoc dataset structure, count the number of disease classes, and determine the total number of training images.

## Theory
The PlantDoc dataset is organized in a folder-based structure where each disease class has its own directory containing sample images. By using `os.listdir()` and recursive directory traversal, we can:
- Identify all disease classes
- Count total images per class
- Understand the dataset composition

This foundational analysis helps us understand the dataset before preprocessing.

```python
import os
from pathlib import Path

# Dataset path
dataset_root = "/Users/nishant/Documents/vscode/plant_disease_detection/data/PlantDoc-Dataset-master"
train_path = os.path.join(dataset_root, "train")

print("="*60)
print("PLANTDOC DATASET ANALYSIS")
print("="*60)

# Check if path exists
if os.path.exists(train_path):
    print(f"\n✓ Dataset found at: {train_path}")
else:
    print(f"\n✗ Dataset NOT found at: {train_path}")
    print("Please verify the dataset path.")
    raise FileNotFoundError(f"Dataset path not found: {train_path}")

# Get all class directories
class_dirs = [d for d in os.listdir(train_path) if os.path.isdir(os.path.join(train_path, d))]
class_dirs.sort()

print(f"\n📊 DATASET STATISTICS:")
print(f"Total Classes: {len(class_dirs)}")

# Count images per class
total_images = 0
class_counts = {}

print(f"\n📁 CLASS DISTRIBUTION:")
print("-" * 60)

for idx, class_name in enumerate(class_dirs, 1):
    class_path = os.path.join(train_path, class_name)
    images = [f for f in os.listdir(class_path) 
              if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
    image_count = len(images)
    class_counts[class_name] = image_count
    total_images += image_count
    print(f"{idx:2d}. {class_name:40s} : {image_count:4d} images")

print("-" * 60)
print(f"\n📈 TOTAL IMAGES: {total_images}")
print(f"📊 AVERAGE IMAGES PER CLASS: {total_images // len(class_dirs)}")
print("\n" + "="*60)
```

## Result
The analysis successfully identified all disease classes in the PlantDoc dataset training folder. The dataset contains multiple plant disease categories with varying numbers of images per class. This information is crucial for understanding the dataset composition and potential class imbalance.

## Conclusion
The PlantDoc dataset is well-organized with clear class separation. Understanding the total count of classes and images provides the foundation for subsequent preprocessing and training experiments.
