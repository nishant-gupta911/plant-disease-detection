# Experiment 2: Data Preprocessing Pipeline

## Aim
To define and demonstrate the image preprocessing pipeline using torchvision transforms for consistent image standardization.

## Theory
Image preprocessing is essential for deep learning models. Common preprocessing steps include:
- **Resizing**: Standardize all images to uniform dimensions (224x224 pixels)
- **Tensor Conversion**: Convert PIL images to PyTorch tensors
- **Normalization**: Apply ImageNet normalization (mean and std)
- **Augmentation**: Apply transformations for training robustness

torchvision.transforms provides efficient, GPU-accelerated transformations.

```python
import os
from PIL import Image
from torchvision import transforms
import torch
import matplotlib.pyplot as plt
import numpy as np

# Define transforms
print("="*70)
print("IMAGE PREPROCESSING PIPELINE")
print("="*70)

# Training transforms
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Validation transforms (no augmentation)
val_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

print("\n✓ Training Transform Pipeline:")
print(train_transforms)

print("\n✓ Validation Transform Pipeline:")
print(val_transforms)

# Test on a sample image
dataset_root = "/Users/nishant/Documents/vscode/plant_disease_detection/data/PlantDoc-Dataset-master"
train_path = os.path.join(dataset_root, "train")

# Get first class and first image
class_name = os.listdir(train_path)[0]
class_path = os.path.join(train_path, class_name)
image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]

if image_files:
    image_path = os.path.join(class_path, image_files[0])
    print(f"\n📂 Testing on sample image:")
    print(f"   Class: {class_name}")
    print(f"   Image: {image_files[0]}")
    
    # Load image
    img = Image.open(image_path)
    print(f"\n🔍 Original Image Properties:")
    print(f"   Size: {img.size}")
    print(f"   Mode: {img.mode}")
    
    # Apply transforms
    transformed_img = train_transforms(img)
    print(f"\n✓ Transformed Image Properties:")
    print(f"   Shape: {transformed_img.shape}")
    print(f"   Data Type: {transformed_img.dtype}")
    print(f"   Min Value: {transformed_img.min():.4f}")
    print(f"   Max Value: {transformed_img.max():.4f}")
    print(f"   Mean: {transformed_img.mean():.4f}")
    print(f"   Std: {transformed_img.std():.4f}")

print("\n" + "="*70)
print("✓ Preprocessing pipeline defined and tested successfully!")
print("="*70)
```

## Result
The preprocessing pipeline was successfully created using torchvision transforms. Images are resized to 224x224 pixels, converted to tensors, and normalized using ImageNet statistics (mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]). The transformed tensors have values normalized around 0 with unit variance.

## Conclusion
The standardized preprocessing pipeline ensures all images are in the correct format for deep learning models. Consistent image preprocessing improves model convergence and performance across different input images.
