```python
import pandas as pd, matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/PlantDisease/logs/training_log.csv')
p1 = df[df['phase']=='P1']
p2 = df[df['phase']=='P2']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

ax1.plot(p1['epoch'], p1['val_acc'], 'b--', label='P1 Val')
ax1.plot(p2['epoch'], p2['tr_acc'],  'g-',  label='P2 Train')
ax1.plot(p2['epoch'], p2['val_acc'], 'r-',  label='P2 Val', linewidth=2)
ax1.axvline(x=p1['epoch'].max(), color='orange', linestyle=':', label='Phase switch')
ax1.set_title('Accuracy'); ax1.legend(); ax1.set_xlabel('Epoch')

ax2.plot(p1['epoch'], p1['val_loss'], 'b--', label='P1 Val Loss')
ax2.plot(p2['epoch'], p2['tr_loss'],  'g-',  label='P2 Train Loss')
ax2.plot(p2['epoch'], p2['val_loss'], 'r-',  label='P2 Val Loss')
ax2.axvline(x=p1['epoch'].max(), color='orange', linestyle=':')
ax2.set_title('Loss'); ax2.legend(); ax2.set_xlabel('Epoch')

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/PlantDisease/logs/training_curves.png', dpi=150)
plt.show()
```
